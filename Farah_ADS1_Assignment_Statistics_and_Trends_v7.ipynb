```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Ignore warnings
warnings.filterwarnings("ignore")
```

```python
# Load the datasets
df_residents_65plus = pd.read_csv('DataGov_Residents_65Plus.csv', index_col='year')

# Print out the dataframe
print("Residents 65Plus Data:")
print(df_residents_65plus)

# Load the datasets
df_nursinghome = pd.read_csv('DataGov_NursingHome_Beds.csv', index_col='year')

# Print out the dataframe
print("\nNursing Home Data:")
print(df_nursinghome)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data
residents = [340700, 370300, 389800, 413100, 466300, 460900, 492700, 535600, 556300, 605000, 622900, 658900, 683800]
beds = [9551, 9690, 9721, 10838, 10968, 12185, 13022, 14918, 15205, 16059, 16221, 17064, 18029]
years = list(range(2010, 2023))

# Calculate beds per 1000 residents
beds_per_1000 = [round((beds[i] / residents[i] * 1000), 2) for i in range(len(residents))]

# Create plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Simple Scatter Plot
ax1.scatter(residents, beds, c=years, cmap='viridis', s=60)
ax1.set_xlabel('Residents 65+')
ax1.set_ylabel('Nursing Home Beds')
ax1.set_title('Residents vs Nursing Home Beds\n(2010-2022)')
ax1.grid(True, alpha=0.3)

# Add year labels
for i, year in enumerate(years):
    ax1.annotate(str(year), (residents[i], beds[i]), xytext=(5, 5),
                textcoords='offset points', fontsize=8)

# Simple Bubble Plot
bubble_sizes = [size * 15 for size in beds_per_1000]  # Scale for visibility
scatter = ax2.scatter(residents, beds, s=bubble_sizes, c=years, cmap='plasma', alpha=0.6)
ax2.set_xlabel('Residents 65+')
ax2.set_ylabel('Nursing Home Beds')
ax2.set_title('Bubble Plot: Size = Beds per 1000 Residents')
ax2.grid(True, alpha=0.3)

# Add year labels
for i, year in enumerate(years):
    ax2.annotate(str(year), (residents[i], beds[i]), xytext=(5, 5),
                textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()

# Print results
print("Yearly Data:")
for i, year in enumerate(years):
    print(f"{year}: {residents[i]:,} residents, {beds[i]:,} beds, {beds_per_1000[i]:.2f} beds/1000")
```

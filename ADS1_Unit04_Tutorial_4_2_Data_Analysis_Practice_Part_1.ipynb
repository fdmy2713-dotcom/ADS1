```python
# Step 1 : import library
# Importing the initial required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To ignore the warnings that may appear
import warnings
warnings.filterwarnings("ignore")
```

```python
# Step 2 : import data from https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv
# Load in the data from the following URL
url = "https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv"
df_salary = pd.read_csv(url)

#print out the dataframe
print(df_salary)
```

```python
# Check and see if you can see the first 5 rows of the dataset
print(df_salary.head())
```

```python
#create a scatter plot
# Scatter plot: Years of Experience vs Salary
plt.scatter(df_salary['Experience Years'], df_salary['Salary'], color='blue')

# Add title and labels
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

# Show the plot
plt.show()
```

```python
# Exploratory Data Analysis (EDA)
# Find out how many rows do you have in this dataset?

# Number of rows and columns
print("Shape of the dataset:", df_salary.shape)

# Number of rows only
print("Number of rows:", df_salary.shape[0])
```

```python
# what is the min-max range of the salary in this dataset?
# what is the average experience years and the average of salary?
# print out the answers

# Minimum and Maximum salary
min_salary = df_salary['Salary'].min()
max_salary = df_salary['Salary'].max()

# Average Years of Experience
avg_experience = df_salary['Experience Years'].mean()

# Average Salary
avg_salary = df_salary['Salary'].mean()

# Print the results
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Average Years of Experience: {avg_experience:.2f}")
print(f"Average Salary: {avg_salary:.2f}")
```

```python
# What is the correlation between Experience years and Salary?
# Is it a strong correlation?

# Calculate correlation between YearsExperience and Salary
correlation = df_salary['Experience Years'].corr(df_salary['Salary'])

# Print the correlation
print(f"Correlation between Years of Experience and Salary: {correlation:.2f}")
```

```python
#find out more information about the data set
# Understanding data in different ways

# Find out the dimension of the dataset
print("Dimension of the data:")
print(df_salary.shape)  # returns (rows, columns)

# Find out the summary of the dataset
print("Summary of the data\n")
print(df_salary.info())  # shows column types, non-null counts

#Find out if there are any missinng values
print("Missing values in the entire dataframe\n")
print(df_salary.isnull().sum())  # sums missing values column-wise

#Find out the statistical decription of the dataset
print("Statistical description:")
print(df_salary.describe())  # min, max, mean, std, quartiles
```

Show the heat map

```python
#Show the heatmap using Matplotlib method

# Compute the correlation matrix
corr_matrix = df_salary.corr()

# Create the heatmap
plt.figure(figsize=(6,5))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()  # Show color scale

# Add labels
plt.xticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns)
plt.yticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns)

# Add correlation values on the heatmap
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}", ha='center', va='center', color='white')

plt.title("Correlation Heatmap")
plt.show()
```

```python
# show the heat map using Seaborn library

# Compute the correlation matrix
corr_matrix = df_salary.corr()

# Create the heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

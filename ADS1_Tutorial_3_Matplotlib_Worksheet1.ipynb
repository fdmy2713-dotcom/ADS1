## Exercise 4 â€” Scatter Plot & Markers
Generate 50 random points around a line `y = 2x + 1` with some noise.

## Exercise 6 â€” Histogram
Generate 1000 samples from a normal distribution with mean 0 and std 1.

**Goals:**
- Plot a histogram with 30 bins
- Show density (probability) instead of counts (`density=True`)
- Add a vertical line at the mean


````python
# TODO: Histogram with density
samples = rng.normal(0, 1, size=1000)

# Your code here
# plt.hist(samples, bins=30, density=True)
# plt.axvline(samples.mean(), linestyle='--')
# plt.title('Normal(0,1) Histogram')
# plt.show()

# TODO: Histogram with density
plt.hist(samples, bins=30, density=True, color="skyblue", edgecolor="black", alpha=0.7)

# Add vertical line at the mean
plt.axvline(samples.mean(), color="red", linestyle="--", linewidth=2, label=f"Mean = {samples.mean():.2f}")

plt.title("Normal(0,1) Histogram")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
````

## Exercise 7 â€” Subplots
Create a figure with **two subplots (1 row, 2 columns)**:
- Left: `sin(x)`
- Right: `cos(x)`

**Goals:**
- Use `plt.subplots(1, 2, figsize=(10,4))`

````python
# TODO: Subplots of sin and cos
x = np.linspace(0, 2*np.pi, 200)
ysin = np.sin(x)
ycos = np.cos(x)

# fig, axes = plt.subplots(1, 2, figsize=(10, 4))
# ax1, ax2 = axes
# ax1.plot(...); ax1.set_title('sin(x)')
# ax2.plot(...); ax2.set_title('cos(x)')
# plt.show()

# TODO: Subplots of sin and cos
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

ax1, ax2 = axes
ax1.plot(x, ysin, color="blue", linewidth=2)
ax1.set_title("sin(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("sin(x)")
ax1.grid(True)

ax2.plot(x, ycos, color="red", linestyle="--", linewidth=2)
ax2.set_title("cos(x)")
ax2.set_xlabel("x")
ax2.set_ylabel("cos(x)")
ax2.grid(True)

plt.tight_layout()
plt.show()
````

## Exercise 10 â€” Fill Between (Confidence Band)
Simulate a noisy signal and a 1-Ïƒ confidence band.

**Goals:**
- Create `x` from 0 to 10 (200 points)
- Let `y = sin(x) + noise` with noise `~ Normal(0, 0.2)`
- Let `sigma = 0.3`
- Plot the line and fill the area between `y - sigma` and `y + sigma` using `plt.fill_between`


````python
# TODO: Plot a line with a confidence band using fill_between
x = np.linspace(0, 10, 200)
noise = rng.normal(0, 0.2, size=x.size)
y = np.sin(x) + noise
sigma = 0.3

# Your code here
# plt.plot(x, y, label='signal')
# plt.fill_between(x, y - sigma, y + sigma, alpha=0.3, label='Â±1Ïƒ')
# plt.legend()
# plt.show()

# TODO: Plot a line with a confidence band using fill_between
plt.plot(x, y, label="signal", color="blue")
plt.fill_between(x, y - sigma, y + sigma, alpha=0.3, color="blue", label="Â±1Ïƒ")

plt.title("Signal with Confidence Band")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
````

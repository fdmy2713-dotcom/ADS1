```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generating a larger dataset with continuous values
np.random.seed(42) # For reproducibility
data_continuous = np.random.normal(loc=20, scale=5, size=1000) # Mean 20, std 5

# Creating a second dataset using pandas
data = {
  'Category': ['A', 'B', 'C', 'D', 'E'],
  'Values1': [10, 20, 15, 30, 25],
  'Values2': [12, 18, 22, 28, 35],
  'Values3': [5, 15, 10, 25, 20]
}

df = pd.DataFrame(data)

# Displaying the dataset

print("Sample Dataset:")
print(df)
```

```python
#Creating a Histogram
sns.histplot(data_continuous, bins=30, kde=True, color='darkolivegreen')
```

```python
#Creating a Single Box Plot
sns.boxplot(data=data_continuous, color='darkolivegreen')
```

```python
#Plotting the simple bar chart using seaborn
sns.barplot(x='Category', y='Values1', data=df, palette='Oranges', hue='Category')
```

```python
## Creating a stacked bar chart
df.set_index('Category')[['Values1', 'Values2', 'Values3']].plot(kind='bar', stacked=True)
plt.title("Stacked Bar Chart using Pandas")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend(title="Values")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

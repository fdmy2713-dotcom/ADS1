```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To ignore the warnings that may appear
import warnings
warnings.filterwarnings("ignore")
```

```python
# Relational Graph (Scatter Plot)
# Showing the correlation between Age and Health Expenditure.

# Function Name
def plot_relational_graph(df):
    """
    Relational Graph: Scatter plot showing relationship between Age and Healthcare Expenditure.
    """
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=df, x='Age', y='Healthcare_Expenditure', alpha=0.7)
    sns.regplot(data=df, x='Age', y='Healthcare_Expenditure', scatter=False, color='red')

    # Add Title and Labels
    plt.title('Relationship between Age and Healthcare Expenditure')
    plt.xlabel('Age')
    plt.ylabel('Healthcare Expenditure ($)')
    plt.grid(True)

    # Show the Plot
    plt.show()
```

```python
# Relational Graph (Scatter Plot)

# Create a sample DataFrame for demonstration
data = {'Age': [25, 35, 45, 55, 65, 75, 85],
        'Healthcare_Expenditure': [1000, 2000, 3500, 5000, 7000, 9000, 12000]}
df = pd.DataFrame(data)

# Now call the function with the created DataFrame
plot_relational_graph(df)
```

```python
# Categorical Graphs (Stacked Bar)
# Stacked bar chart showing usage of care services by Age group.

# Function Name
def plot_categorical_graph(df):
    """
    Categorical Graph: Stacked bar chart showing usage of care services by Age group.
    """
    df_pivot = df.pivot_table(values='Usage_Count', index='Age_Group', columns='Service_Type', aggfunc='sum')
    df_pivot.plot(kind='bar', stacked=True, figsize=(8,6))

    # Add Title and Labels
    plt.title('Usage of Long-Term Care Services by Age Group')
    plt.xlabel('Age Group')
    plt.ylabel('Number of Users')
    plt.legend(title='Service Type')

    # Show the Plot
    plt.show()
```

```python
# Categorical Graphs (Stacked Bar)

# Create a sample DataFrame for categorical graph demonstration
data_categorical = {'Age_Group': ['20-30', '20-30', '30-40', '30-40', '60-70', '60-70', '80+'],
                    'Service_Type': ['Home Care', 'Nursing Home', 'Home Care', 'Nursing Home', 'Home Care', 'Nursing Home', 'Nursing Home'],
                    'Usage_Count': [10, 5, 15, 8, 20, 12, 18]}
df_categorical = pd.DataFrame(data_categorical)

# Now call the function with the created categorical DataFrame
plot_categorical_graph(df_categorical)
```

```python
# Categorical Graphs (Pie Chart)
# Proportion of causes of hospitalization among seniors.

# Function Name
def plot_categorical_pie(df):
    """
    Pie Chart: Proportion of causes of hospitalization among seniors.
    """
    df_group = df.groupby('Cause')['Count'].sum()
    df_group.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6))

    # Add Title and Labels
    plt.title('Causes of Hospitalization Among Seniors')
    plt.ylabel('')

    # Show the Plot
    plt.show()
```

```python
# Categorical Graphs (Pie Chart)

# Create a sample DataFrame for categorical pie chart demonstration
data_pie = {'Cause': ['Heart Disease', 'Stroke', 'Pneumonia', 'Diabetes', 'Other'],
            'Count': [25, 15, 10, 8, 42]}
df_pie = pd.DataFrame(data_pie)

# Now call the function with the created pie chart DataFrame
plot_categorical_pie(df_pie)
```

```python
# Statistical Graphs (Line Chart)
# Line chart showing trend of Life Expectancy at Age 65 over time.

# Function Name
def plot_statistical_graph(df):
    """
    Statistical Graph: Line chart showing trend of Life Expectancy at Age 65 over time.
    """
    plt.figure(figsize=(8,6))
    sns.lineplot(data=df, x='Year', y='Life_Expectancy_65', marker='o')
    plt.title('Trend of Life Expectancy at Age 65 (Singapore)')
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy at Age 65 (Years)')
    plt.grid(True)

    # Show the Plot
    plt.show()
```

```python
# Statistical Graphs (Line Chart)

# Create a sample DataFrame for statistical graph demonstration
data_statistical = {'Year': [2000, 2005, 2010, 2015, 2020],
                    'Life_Expectancy_65': [18.5, 19.2, 20.1, 20.9, 21.5]}
df_statistical = pd.DataFrame(data_statistical)

# Now call the function with the created statistical DataFrame
plot_statistical_graph(df_statistical)
```

```python
# Statistical Graphs (Box Plot)
# Statistical distribution of Healthcare Costs by Age Group.

# Function Name
def plot_statistical_box(df):
    """
    Box Plot: Statistical distribution of Healthcare Costs by Age Group.
    """
    plt.figure(figsize=(8,6))
    sns.boxplot(data=df, x='Age_Group', y='Healthcare_Expenditure')

    # Add Title and Labels
    plt.title('Distribution of Healthcare Expenditure by Age Group')
    plt.xlabel('Age Group')
    plt.ylabel('Healthcare Expenditure ($)')

    # Show the Plot
    plt.show()
```

```python
# Statistical Graphs (Box Plot)

# Create a sample DataFrame for statistical box plot demonstration
data_box = {'Age_Group': ['20-30', '20-30', '30-40', '30-40', '60-70', '60-70', '80+', '80+'],
            'Healthcare_Expenditure': [1500, 2500, 3000, 4000, 6000, 8000, 10000, 15000]}
df_box = pd.DataFrame(data_box)

# Now call the function with the created statistical DataFrame
plot_statistical_box(df_box)
```
